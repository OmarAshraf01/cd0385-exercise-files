# Exercise: Iris Dataset
Now that you have a good understanding of exploritory data analysis and it's importance, it's time to put your knowledge to a more practical example. We'll be focusing on a iris public dataset from the scikit-learn library.

Our main objectives for this dataset are:
1. Load the iris dataset into a pandas dataframe
2. Create a table summary of the features and target values
3. Create a histogram of all the features and target
4. Create a correlation matrix of the features and target
5. Create scatter plots of all the features and target

```python
import pandas as pd
import sklearn
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
# Load in the iris dataset
iris = datasets.load_iris()
```

```python
# For clarity, the iris dataset is a dictionary with the data and target separated
iris.keys()
```

```python
# Create the iris `data` dataset as a dataframe and name the columns with `feature_names`
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])

# Include the target as well
df['target'] = iris['target']
```

# Scatter Plot Of Features
Create a scatter plot of the four features against eachother to visualize the results from the correlation matrix
1. `sepal length (cm)` vs. `sepal width (cm)`
2. `sepal length (cm)` vs. `petal length (cm)`
3. `sepal length (cm)` vs. `petal width (cm)`
4. `sepal width (cm)` vs. `petal length (cm)`
5. `sepal width (cm)` vs. `petal width (cm)`
6. `petal length (cm)` vs. `petal width (cm)`

```python
# Create a figure with subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Scatter plots
df.plot.scatter(x='sepal length (cm)', y='sepal width (cm)', ax=axes[0, 0], color='blue', alpha=0.6)
df.plot.scatter(x='sepal length (cm)', y='petal length (cm)', ax=axes[0, 1], color='green', alpha=0.6)
df.plot.scatter(x='sepal length (cm)', y='petal width (cm)', ax=axes[0, 2], color='red', alpha=0.6)
df.plot.scatter(x='sepal width (cm)', y='petal length (cm)', ax=axes[1, 0], color='purple', alpha=0.6)
df.plot.scatter(x='sepal width (cm)', y='petal width (cm)', ax=axes[1, 1], color='orange', alpha=0.6)
df.plot.scatter(x='petal length (cm)', y='petal width (cm)', ax=axes[1, 2], color='brown', alpha=0.6)

# Set titles for each subplot
axes[0, 0].set_title('Sepal Length vs Sepal Width')
axes[0, 1].set_title('Sepal Length vs Petal Length')
axes[0, 2].set_title('Sepal Length vs Petal Width')
axes[1, 0].set_title('Sepal Width vs Petal Length')
axes[1, 1].set_title('Sepal Width vs Petal Width')
axes[1, 2].set_title('Petal Length vs Petal Width')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

# Scatter Plot Of Features And Target
Create a scatter plot of the four features against the target
1. `sepal length (cm)`
2. `sepal width (cm)`
3. `petal length (cm)`
4. `petal width (cm)`

```python
# Create a figure with subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Scatter plots
scatter = axes[0, 0].scatter(df['sepal length (cm)'], df['target'], c=df['target'], cmap='viridis')
axes[0, 0].set_title('Sepal Length vs Target')
axes[0, 0].set_xlabel('Sepal Length (cm)')
axes[0, 0].set_ylabel('Target')

scatter = axes[0, 1].scatter(df['sepal width (cm)'], df['target'], c=df['target'], cmap='viridis')
axes[0, 1].set_title('Sepal Width vs Target')
axes[0, 1].set_xlabel('Sepal Width (cm)')
axes[0, 1].set_ylabel('Target')

scatter = axes[1, 0].scatter(df['petal length (cm)'], df['target'], c=df['target'], cmap='viridis')
axes[1, 0].set_title('Petal Length vs Target')
axes[1, 0].set_xlabel('Petal Length (cm)')
axes[1, 0].set_ylabel('Target')

scatter = axes[1, 1].scatter(df['petal width (cm)'], df['target'], c=df['target'], cmap='viridis')
axes[1, 1].set_title('Petal Width vs Target')
axes[1, 1].set_xlabel('Petal Width (cm)')
axes[1, 1].set_ylabel('Target')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
